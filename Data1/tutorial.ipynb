```python
import glob
import os
import numpy as np
import matplotlib.pyplot as plt

files=glob.glob('*.csv')
print(files)
fig = plt.figure(figsize=(4,3))
for i in files:
    data=np.loadtxt(i,delimiter=None)
    x=data[:,0]
    y=data[:,1]
    plt.plot(x,y)
    plt.xlabel('Voltage(V)')
    plt.ylabel('Current(I)')
```
