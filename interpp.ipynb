```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

## sample data
x=np.linspace(0,10,num=11,endpoint=True)
y=np.sin(x)

#create interpolation function
f=interp1d(x,y,kind='cubic')
x_new=np.linspace(0,10,num=100,endpoint='True')
y_new=f(x_new)

fig,ax=plt.subplots()
ax.plot(x,y,'o', label='data')
ax.plot(x_new,y_new,'+', label='interpolated')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import RectBivariateSpline

x=np.linspace(0,20,num=21, endpoint=True)
y=np.linspace(0,20,num=21, endpoint=True)
X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)

x_new=np.linspace(0,20,num=201, endpoint=True)
y_new=np.linspace(0,20,num=201, endpoint=True)

f=RectBivariateSpline(x,y,Z)
Z_new=f(x_new,y_new)

fig,(ax1,ax2)=plt.subplots(ncols=2, figsize=(10,4))
ax1.pcolormesh(X,Y,Z, shading='auto')
ax1.set(title='original data')
ax2.pcolormesh(x_new,y_new,Z_new, shading='auto')
ax2.set(title='interpolated data')
plt.show()
```
